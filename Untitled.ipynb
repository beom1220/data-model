```python
# 데이터 읽어오기
import pandas as pd
df = pd.read_csv('A회사_보습제_매출데이터_v1.csv', encoding='cp949')
print(df)
```

```python
# 6개월간 매출추이 시각화
import matplotlib.pyplot as plt

dates = df.columns[1:]
total_sales = df.loc[0, dates].str.replace('$', '').str.replace(',', '').astype(float)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(dates, total_sales, marker='o', linestyle='-', color='blue')
plt.title('Total sales trend for 6 months', fontsize=14)
plt.xlabel('month', fontsize=12)
plt.ylabel('sales (dollar)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 광고 비용 시각화

monthly_costs = df.loc[2, df.columns[1:]].str.replace('[^\d.]', '', regex=True).astype(float)

# '월' 열 추가
monthly_costs.index = df.columns[1:]

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(monthly_costs.index, monthly_costs.values, marker='o', linestyle='-', color='green')
plt.title('month-cost', fontsize=14)
plt.xlabel('month', fontsize=12)
plt.ylabel('cost', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 매출, 목표매출, 광고비용, SNS광고비용 시각화

dates = df.columns[1:]
total_sales = df.loc[0, dates].str.replace('$', '').str.replace(',', '').astype(float)
goal_sales = df.loc[1, dates].str.replace('$', '').str.replace(',', '').astype(float)
monthly_costs = df.loc[2, dates].str.replace('[^\d.]', '', regex=True).astype(float)
sns_ad_costs = df.loc[3, dates].str.replace('[^\d.]', '', regex=True).astype(float)

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 그래프: 6개월간 매출추이
plt.plot(dates, total_sales, marker='o', linestyle='-', color='blue', label='Total Sales')

# 그래프 : 목표 매출
plt.plot(dates, goal_sales, marker='o', linestyle='-', color='green', label='Total Sales')

# 그래프: 월별 광고 비용
plt.plot(dates, monthly_costs.values, marker='o', linestyle='-', color='orange', label='Ad Costs')

# 그래프: SNS 광고 비용
plt.plot(dates, sns_ad_costs.values, marker='o', linestyle='-', color='magenta', label='SNS Ad Costs')

plt.title('Sales, Adjusted Costs, and SNS Ad Costs', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.show()
```

```python
# 매출, 목표매출, 광고비용, SNS광고비용 시각화

dates = df.columns[1:]
total_sales = df.loc[0, dates].str.replace('$', '').str.replace(',', '').astype(float)
goal_sales = df.loc[1, dates].str.replace('$', '').str.replace(',', '').astype(float)
monthly_costs = df.loc[2, dates].str.replace('[^\d.]', '', regex=True).astype(float)
sns_ad_costs = df.loc[3, dates].str.replace('[^\d.]', '', regex=True).astype(float)

# 두 그래프를 하나의 fig에 그리기
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
fig.subplots_adjust(hspace = 0.1)

# 그래프: 6개월간 매출추이
ax1.plot(dates, total_sales, marker='o', linestyle='-', color='blue', label='Total Sales')
# 그래프 : 목표 매출
ax1.plot(dates, goal_sales, marker='o', linestyle='-', color='green', label='Goal Sales')

# 그래프: 월별 광고 비용
ax2.plot(dates, monthly_costs.values, marker='o', linestyle='-', color='orange', label='Ad Costs')
# 그래프: SNS 광고 비용
ax2.plot(dates, sns_ad_costs.values, marker='o', linestyle='-', color='magenta', label='SNS Ad Costs')

ax1.set_ylim(5000000, 7000000) # 윗 부분 y축 범위 설정
ax2.set_ylim(0, 1250000) # 아랫 부분 y축 범위 설정
ax1.grid(True)
ax2.grid(True)

# 두 그래프 사이의 경계선 제거
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)
ax2.xaxis.tick_bottom()
# 두 그래프 사이의 y축에 물결선 효과 마커 표시
kwargs = dict(marker=[(-1, -0.5), (1, 0.5)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

fig.suptitle('Sales, Adjusted Costs, and SNS Ad Costs', fontsize=14)
ax2.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Amount', fontsize=12)
plt.xticks(rotation=45)
ax1.legend()
ax2.legend()

plt.show()
```
